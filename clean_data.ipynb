## Clean GPT dataset

```python
df[df['instruction'].str.startswith('[')]
```

```python
example = df['instruction'][9632]
example
```

```python
def clean_data(text):
    result = re.sub(r'\[.*\]\n\n', '', text)
    result = re.sub(r'\[.*\]\n', '', result)
    result = re.sub(r'\[.*\]', '', result)
    result = re.sub(r'Patient.*:\n', '', result)
    result = re.sub(r'Patient.*:', '', result)
    return result
```

```python
df['instruction'] = df['instruction'].apply(clean_data)
```

```python
df.to_csv('/cbica/home/xjia/qlora/data/lab/self_instruct_gpt3.5_clean.csv', index=False)
```
