```python
import pandas as pd
import json
```

```python
jt = pd.read_json('/cbica/home/xjia/qlora/data/lab/JT_bench.jsonl', lines=True)
jt
```

```python
jt["turns"] = [[q] for q in jt["turns"]]
jt
```

```python
mhmc = pd.read_csv('/cbica/home/xjia/qlora/data/lab/mental_health_exam_MC.csv')
mhmc
```

```python
mhmc.rename(columns={'answer': 'answer_full'}, inplace=True)

mhmc["answer"] = mhmc["answer_full"].apply(lambda x: x.strip()[0].upper())
mhmc
```

```python
mhmc.drop(columns=['answer_full'], inplace=True)

new_mhmc = pd.concat([mhmc, jt], ignore_index=True)
new_mhmc
```

```python
new_mhmc["answer"].value_counts()
```

```python
for index, row in new_mhmc.iterrows():
    with open('/cbica/home/xjia/qlora/data/lab/JT_bench_new.jsonl', 'a') as fout:
            json.dump({
                "question_id": row["question_id"],
                "category": row["category"],
                "turns": row["turns"],
                "answer": row["answer"],
            }, fout)
            fout.write('\n')
```

```python
evalq = pd.read_csv('/cbica/home/xjia/qlora/data/lab/EvalQ_Cleaned_New.csv')
evalq['length'] = evalq["Questions"].apply(lambda x: len(x.strip().split()))
evalq = evalq[evalq["length"] > 50]
evalq
```

```python
evalq_mini = evalq.sample(200, random_state=42)
evalq_mini.reset_index(drop=True, inplace=True)
evalq_mini
```

```python
output_path = '/cbica/home/xjia/qlora/data/lab/evalq_bench.jsonl'

for index, row in evalq_mini.iterrows():
    with open(output_path, 'a') as fout:
        json.dump({
            "question_id": index+1,
            "category": "generic",
            "turns": [row["Questions"]]
        }, fout)
        fout.write('\n')
```

```python
evalq_mini["instruction"] = "You are a helpful mental health counselling assistant, please answer the mental health questions based on the patient's description.\nThe assistant gives helpful, comprehensive, and appropriate answers to the user's questions."
evalq_mini.rename(columns={'Questions': 'input'}, inplace=True)
evalq_mini
```

```python
evalq_mini[["instruction", "input"]].to_csv('/cbica/home/xjia/qlora/data/lab/evalq_bench.csv', index=False)
```
